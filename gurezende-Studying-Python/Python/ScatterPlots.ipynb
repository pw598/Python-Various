# Scatter Plots

Scatterplot (or points graphic) is the type of graphic where each point is represented by a coordinate in X and Y axes.
Those are one of the best ways to show relationship between two numeric variables.

```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating a dataframe
df = pd.DataFrame({'A':[1,10,15,34,22,44,22,32,32,12,11,22,44,55,63,34,23,45,22,44,22,33,32,11,12,32,21],
                   'B':[2,22,31,60,44,60,50,60,67,30,30,40,70,77,100,50,44,80,44,88,40,60,65,23,25,55,44]})
```

```python
# Show first rows of the data
df.head()
```

```python
# Plot your scatterplot
sns.scatterplot(x='A', y='B', data=df, color = 'forestgreen')
plt.show()
```

Notice how you can draw a line with your eyes to represent the relationship between both variables. This shows a strong linear correlation between A and B.

```python
# Using jointplot, you can extract even more information. Try it using the parameter kind = 'reg' to show the regression line.
sns.jointplot(x= 'A', y= 'B', data=df, kind='reg', height=8)
plt.xlim(-5,80)
plt.ylim(-5,120)
plt.show()
```

```python
# Calculate the Correlation between A and B: 95%
df.corr().style.background_gradient(cmap='coolwarm')
```

```python
# Example of variables with low correlation
df2 = pd.DataFrame({'C':np.random.randint(1,100,50), 'D':np.random.randint(1,100,50)})
sns.scatterplot(x='C', y='D', data=df2)
plt.show()
```

```python
# Correlation in df2: only 16%
df2.corr()
```
