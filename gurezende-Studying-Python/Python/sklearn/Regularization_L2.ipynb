## Visualizing effect of alpha in Ridge Regularization
The higher the alpha, the more the model is regularized and the regression line is going to be flatter, going towards the dataset's mean value.
See the slope in the graphic.

```
import scipy.stats as scs
# Create Dataset
X = scs.norm.rvs(0,0.1,size=100)-20
y= X * scs.norm.rvs(1,40, size=100) + 1
```

```
# Set figure
plt.figure(figsize=(20,7))
#Plot 1
# Fit model
dummy_ridge = Ridge(alpha=0.001).fit(X.reshape(-1, 1), y)
preds = dummy_ridge.predict(X.reshape(-1, 1))
plt.subplot(121)
sns.scatterplot(x=X, y=y, alpha=.8)
sns.lineplot(x=X, y=preds, color='red')
plt.title(f'alpha = 0.001 | slope = {dummy_ridge.coef_}');

#Plot 2
# Fit model
dummy_ridge = Ridge(alpha=1000).fit(X.reshape(-1, 1), y)
preds = dummy_ridge.predict(X.reshape(-1, 1))
plt.subplot(122)
sns.scatterplot(x=X, y=y, alpha=.8)
sns.lineplot(x=X, y=preds, color='red')
plt.title(f'alpha = 1000 | slope = {dummy_ridge.coef_}');
```
